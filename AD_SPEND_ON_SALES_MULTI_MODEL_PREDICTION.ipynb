```python
df["TV"]=df["TV"].astype(int)
df["radio"]=df["radio"].astype(int)
df["newspaper"]=df["newspaper"].astype(int)
df["sales"]=df["sales"].astype(int)
```

```python
x=df.iloc[:,:3]
x
```

```python
y=df.iloc[:,3]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model_1 = LinearRegression()
model_2 = SGDRegressor()
model_3 = Lasso()
model_4 = Ridge()
model_5 = ElasticNet()
model_6 = SVR()
model_7 = KNeighborsRegressor()
model_8 = DecisionTreeRegressor()
model_9 = RandomForestRegressor()
model_10 = BaggingRegressor()
model_11 = ExtraTreesRegressor()
model_12 = AdaBoostRegressor()
model_13 = XGBRegressor(verbose=0)
model_14 = LGBMRegressor(verbose=0)
model_15= MLPRegressor()
```

```python
print(model_1.fit(x_train, y_train))
print(model_2.fit(x_train, y_train))
print(model_3.fit(x_train, y_train))
print(model_4.fit(x_train, y_train))
print(model_5.fit(x_train, y_train))
print(model_6.fit(x_train, y_train))
print(model_7.fit(x_train, y_train))
print(model_8.fit(x_train, y_train))
print(model_9.fit(x_train, y_train))
print(model_10.fit(x_train, y_train))
print(model_11.fit(x_train, y_train))
print(model_12.fit(x_train, y_train))
print(model_13.fit(x_train, y_train))
print(model_14.fit(x_train, y_train))
print(model_15.fit(x_train, y_train))

```

```python
models = [model_1, model_2, model_3, model_4, model_5,
          model_6, model_7, model_8, model_9, model_10,
          model_11, model_12, model_13, model_14,model_15]
models_names = ['LinearRegression', 'SGDRegressor', 'Lasso', 'Ridge', 'ElasticNet', 'SVR', 'KNeighborsRegressor', 'DecisionTreeRegressor', 'RandomForestRegressor', 'BaggingRegressor', 'ExtraTreesRegressor', 'AdaBoostRegressor', 'XGBRegressor', 'LGBMRegressor','MLPRegressor']

# Calculate predictions and squared errors for each model:
squared_errors = []
for model in models:
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    squared_errors.append(f'{mse * 100:.2f}%')  # Format as percentage

train_score = [model.score(x_train, y_train) for model in models]
test_score = [model.score(x_test, y_test) for model in models]

ratio = []
for train, test in zip(train_score, test_score):
    result = train - test
    ratio.append(f'{result * 100:.2f}%')

rate = []
for train, test in zip(train_score, test_score):
    if train <= 0.65 and test <= 0.65:
        rate.append('bad')
    elif train > test * 1.10:
        rate.append('overfite')
    elif train > 0.65 and train < 0.80 and test > 0.65 and test < 0.80:
        rate.append('middle')
    elif train >= 0.80 and test >= 0.80 and train < 1.00 and test < 1.00:
        rate.append('good')
    elif train >= 0.80 and test < 0.80:
        rate.append('high train, low test')
    else:
        rate.append('unknown')

# Create DataFrame
model_score = pd.DataFrame({
    'Model': models_names,
    'Train score': [f'{round(score * 100, 2)}%' for score in train_score],
    'Test score': [f'{round(score * 100, 2)}%' for score in test_score],
    'Ratio difference': ratio,
    'Evaluate model': rate,
    'Squared error': squared_errors
})

model_score
```

```python
best_model_index = model_score['Test score'].idxmax()
best_model_name = model_score.loc[best_model_index, 'Model']
best_model_accuracy = model_score.loc[best_model_index, 'Test score']

print(f"\nBest Model: {best_model_name} with Accuracy: {best_model_accuracy}")
```

```python

best_model_index = models_names.index(best_model_name)

best_model = models[best_model_index]


feature_importances = best_model.feature_importances_


feature_importance_df = pd.DataFrame({
    'Feature': x.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title(f'Feature Importances - {best_model_name}', fontsize=16)
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
